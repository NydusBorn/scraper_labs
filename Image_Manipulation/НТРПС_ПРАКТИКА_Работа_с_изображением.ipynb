## Практика по работе с изображением


# Задание

### Работа с данными изображения

0. Выбрать 2 картинки для задания (можно сделать снимки смартфоном)

0. Цветовые пространства

Изучите файл с цветовыми пространствами.

Конвертируйте исходное изображение из RGB в выбранное цветовое пространство и обратно.

Объясните, для чего используется преобразование цветовых пространств и приведите примеры практического применения.

0. Выделение контуров (Contours)

Реализуйте выделение контуров объектов на изображении.

Определите и визуализируйте найденные контуры.

Опишите, какие методы и параметры использовались для улучшения выделения контуров.

0. Денойзинг (Denoising). Возможно, потребуется сначала добавить шум

Ознакомьтесь с методами шумоподавления.

Примените один или несколько методов денойзинга к изображению с шумом.

Сравните результаты и оцените качество шумоподавления.

0. Морфология (Morphology)

Исследуйте основные морфологические операции по файлу

Реализуйте данные операции на бинарном изображении или маске.

Проанализируйте как морфология помогает улучшить качество выделения объектов и контуров.

0. Преобразования изображений (Image Transformation)

Выполните геометрические преобразования изображения: поворот, масштабирование, сдвиг.

Реализуйте изменение яркости, контраста и фильтрацию изображения.

Опишите как преобразования помогают в предобработке изображений для последующего анализа.
